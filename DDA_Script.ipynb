```python
df_status=df_status[["Survey ID","Status"]].rename(columns={"Status":"Status_New"})
```

```python
df_merge=df_merge.merge(df_status,on="Survey ID",how="left")
```

```python
df_merge.loc[df_merge["Status"]=="Checking","Status"] = df_merge.loc[df_merge["Status"]=="Checking","Status_New"]
```

```python
df_report=df_merge[["Status","Created By"]]
```

```python
df_report.head(2)
```

```python
status_map = {
    "OK": "OK",
    "ok": "OK",
    "PROCESS": "PROCESS",
    "process": "PROCESS",
    "REJECT": "REJECT",
    "Reject": "REJECT",
    "reject": "REJECT",
    "pending": "PENDING"
}

# Strip spaces and map to standard categories
df_report["Status"] = df_report["Status"].str.strip().map(status_map)
```

```python
pivot = pd.crosstab(
    df_report["Created By"],
    df_report["Status"]
)
```
